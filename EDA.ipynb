# Identificación de variables artificiales

Posteriormente al obtener las visualizaciones, optamos por conocer mejor los datos y se hizo una revisión manual rápida, pero efectiva para identificar las variables que tenían pinta de ser artificiales (se seleccionaron las variables que claramente representaban patrones raros), por ejemplo variables que tenían un único valor o variables que tenían dos valores distintos pero solo una insancia tenía un valor diferente.

Las variables que se identificaron como artificiales se capturaron en un bloc de notas (garbae.txt). 

```python
# Obtenemos las variables artificiales del arvhivo txt
garbage_features = preprocessing.get_garbage_features('garbage.txt')
# Eliminamos las variables del dataset
datos =  preprocessing.clean_data(datos, garbage_features)
datos
```

# Variables categóricas

Con respecto a las variables categóricas por cuestiones de tiempo se decidió no tomarlas en cuenta, desde un inició agilizó el entrenamiento de los modelos. La única razón fue esa, por lo tanto hacer el respectivo análisis de las variables categóricas es imortante, ya que podrían mejorar significativamente los resultados.

```python
object_cols = datos.select_dtypes(include=['object']).columns
```

# Selección de variables

Para seleccionar las variables más importantes, se hizo una selección de características usando un modelo lightgbm. 

En lugar de seleccionar las características más importantes del modelo, se decidió eliminar las variables menos significativas para el lightgbm, esto debido a que quedarnos solo con las características seleccionadas por este modelo iba a limitar a los demás modelos, además de no tener un contexto claro sobre el significado de las variables, de esta manera nos aseguramos de perder información y hacer más robustos los modelos de ensamble. 

Además, gracias al análisis visual realizado previamente muchas de las variables menos significativas del lightgbm coincidian con anotaciones de variables candidatas a ser artificiales, lo que respalda esta manera de selección de características.

```python
# División de datos
X = datos.drop(object_cols, axis=1).copy()
X = X.drop(['Target'], axis=1).copy()
y = datos['Target'].copy()
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=43, stratify=y)
# Entrenamiento del modelo
model = models.get_lgbm(X_train, y_train, X_test, y_test)
# Selección de características
importancia = model.feature_importance(importance_type='split')
feature_importance_df = pd.DataFrame({'Feature': X_train.columns, 'Importance': importancia})
feature_importance_df = feature_importance_df.sort_values(by='Importance', ascending=False)
feature_importance_df
```

Seleccionamos las características menos imoprtantes para el modelo.

```python
worst_features_lgbm = feature_importance_df[feature_importance_df['Importance'] == 0 ]
worst_features_lgbm
```

Finalmente obtenemos una lista de la características menos importantes que se utilizaran para entrenar los modelos y servirán para hacer el procesamiento de predicciones de datos nuevos.

```python
# Concatenamos todas las features que no se tomarán en cuenta
worst_features = worst_features_lgbm['Feature'].tolist() + object_cols.tolist() + garbage_features
# Guardamos en un archivo binario
with open('worst_features.pkl', 'wb') as archivo:
    pickle.dump(worst_features, archivo)
```
